```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('EDA_census.csv')
df.head()
```

```python
df[df['Age-group'] == '20-24']['Illiterate Females'].sum()/df[df['Age-group'] == '20-24']['Total Females'].sum()
# df[''
```

```python
df['Literacy Percentage'] = df['Literate Persons']/df['Total Persons']
```

```python
plt.figure(figsize=(23, 5))
sns.barplot(data=df, x='Age-group', y='Literacy Percentage')
```

```python
df['Female Literacy Percentage'] = df['Literate Females']/df['Total Females']
```

```python
plt.figure(figsize=(10, 10))
sns.barplot(data=df, y='Area Name', x='Female Literacy Percentage')
```

```python
plt.figure(figsize=(10, 10))
sns.barplot(data=df, y='Area Name', x='Literacy Percentage')
```

```python
areawise_literate_persons = pd.DataFrame(df.groupby('Area Name')['Literate Persons'].sum()).reset_index()
```

```python
areawise_total_persons = pd.DataFrame(df.groupby('Area Name')['Total Persons'].sum()).reset_index()
```

```python
summary = areawise_total_persons.merge(areawise_literate_persons, on='Area Name')
```

```python
summary['Literacy Percentage'] = summary['Literate Persons']/summary['Total Persons']
```

```python
summary.sort_values('Literacy Percentage')
```
